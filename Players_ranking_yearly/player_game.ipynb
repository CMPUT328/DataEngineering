```python
import json

with open('game_2020.json', 'r') as f:
    game_2020 = json.load(f)

with open('game_2021.json', 'r') as f:
    game_2021 = json.load(f)
    
with open('game_2022.json', 'r') as f:
    game_2022 = json.load(f)
    
with open('game_2023.json', 'r') as f:
    game_2023 = json.load(f)

game = game_2020+game_2021+game_2022+game_2023
```

```python
games = game
```

```python
# add match history to teams
for team in teams_obj:
    team['match_history'] = []
for game in games:
    
    for mapping in mapping_obj:
        
        if mapping["platformGameId"] == game[0]["platformGameId"] and len(mapping["teamMapping"].keys())==2:
            
            for match in matches_obj:
                if match["id"] == mapping["esportsGameId"]:
                    total_time = game[1]["gameTime"]/60000

                    data_15 = game[0]["teams"]

                    data_end = game[1]["teams"]
                    
                    teams = [None,None]

                    for i in range(2):
                        teams[i] = {"Team_id":mapping["teamMapping"][str((i*100)%200 +100)]}
            
                        teams[i]["Opponent_team"] = mapping["teamMapping"][str((i*100 +100)%200 +100)]
                        teams[i]["eventTime"] = game[0]["eventTime"]
                        teams[i]["inhibKills"] = data_end[i]["inhibKills"]
                        teams[i]["LeagueId"] = match["leagueId"]
                        teams[i]["towerKills"] = data_end[i]["towerKills"]
                        
                        teams[i]["baronKills"] = data_end[i]["baronKills"]
                        
                        teams[i]["dragonKills"] = data_end[i]["dragonKills"]
                        
                        teams[i]["Gold_per_min"] = data_end[i]["totalGold"]/total_time
                        
                        teams[i]["KA_per_min"] = (data_end[i]["championsKills"]+data_end[i]["assists"])/total_time

                        teams[i]["deaths_per_min"] = data_end[i]["deaths"]/total_time  
                        
                        teams[i]["game_length"] = total_time
                        
                        teams[i]["Gold_15"] = data_15[i]["totalGold"]
                        
                        teams[i]["towerKills_15"] = data_15[i]["towerKills"]
                        
                        teams[i]["dragonKills_15"] = data_15[i]["dragonKills"]
                        
                        teams[i]["KA_15"] = (data_15[i]["championsKills"]+data_15[i]["assists"])
                        
                        teams[i]["deaths_15"] = data_15[i]["deaths"]
                        
                        for t in match["teams"]:
                                
                            if t["id"] == teams[i]["Team_id"]:
                                    
                                teams[i]["Result"] = 1 if t["result"]["outcome"] == 'win' else 0
                        if 'Result' not in teams[i].keys():
                            teams[i]['Result'] = None
                        
                    for i in range(2):
               
                        teams[i]["GD_15"] = teams[i]["Gold_15"]/(teams[i]["Gold_15"] + teams[(i+1)%2]["Gold_15"])

                        teams[i]["GPMD"] = teams[i]["Gold_per_min"]/(teams[i]["Gold_per_min"] + teams[(i+1)%2]["Gold_per_min"])

                        teams[i]["towerKillsD_15"] = teams[i]["towerKills_15"]/max((teams[(i+1)%2]["towerKills_15"] + teams[i]["towerKills_15"]), 1)

                        teams[i]["dragonKillsD_15"] = teams[i]["dragonKills_15"]/max((teams[(i+1)%2]["dragonKills_15"] + teams[i]["dragonKills_15"]), 1)

                        teams[i]["KAD_15"] = teams[i]["KA_15"]/max((teams[(i+1)%2]["KA_15"] + teams[i]["KA_15"]), 1)

                    for i in range(2):
                        for team in teams_obj:
                            if teams[i]["Team_id"] == team["team_id"] and teams[i]['Result'] != None:
                                team["match_history"].append(teams[i])
                        
with open('teams_record.json', 'w') as f:
    json.dump(teams_obj, f, indent = 4)
            
```

```python
# add match history to players
# change the year at line for game in ... (games for all years, game_2020 for 2020, ...)
for player in players_obj:
    player['match_history'] = []
for game in games:
    
    for mapping in mapping_obj:
        
        if mapping["platformGameId"] == game[0]["platformGameId"]:
            if len(mapping["participantMapping"]) < 10 or len(mapping["teamMapping"]) < 2:
                continue
            for match in matches_obj:
                if match["id"] == mapping["esportsGameId"]:
                    total_time = game[1]["gameTime"]

                    data_15 = game[0]["participants"]

                    data_end = game[1]["participants"]
                    for lane in range(1,6):
                        players = [None,None]

                        for i in range(2):

                            players[i] = {"Player_id":mapping["participantMapping"][str(lane + 5*i)]}
                            players[i]["Lane"] = (lane-1+5*i)%5+1
                            players[i]["LeagueId"] = match["leagueId"]
                            players[i]["eventTime"] = game[0]["eventTime"]
                            players[i]["Player_team"] = mapping["teamMapping"][str(data_15[lane-1 + 5*i]["teamID"])]
                            players[i]["Opponent"] = mapping["participantMapping"][str((lane-1 + 5*i + 5)%10+1)]
                            players[i]["Opponent_team"] = mapping["teamMapping"][str(data_15[(lane-1 + 5*i + 5)%10]["teamID"])]
                            players[i]["Gold_15"] = data_15[lane-1 + 5*i]["totalGold"]

                            players[i]["XP_15"] = data_15[lane-1 + 5*i]["XP"]

                            players[i]["CS_15"] = data_15[lane-1 + 5*i]["stats"][0]["value"] + data_15[lane-1 + 5*i]["stats"][1]["value"]

                            players[i]["CS"] = data_end[lane-1 + 5*i]["stats"][0]["value"] + data_end[lane-1 + 5*i]["stats"][1]["value"]

                            players[i]["CS_per_min"] = round(players[i]["CS"]/total_time*60000, 1)

                            players[i]["Kill"] = data_end[lane-1 + 5*i]["stats"][4]["value"]

                            players[i]["Death"] = data_end[lane-1 + 5*i]["stats"][5]["value"]

                            players[i]["Assist"] = data_end[lane-1 + 5*i]["stats"][6]["value"]

                            players[i]["KDA"] = round(players[i]["Kill"]*players[i]["Assist"]/max(players[i]["Death"],1), 2)

                            players[i]["KA_per_min"] = round((players[i]["Kill"]+players[i]["Assist"])/total_time*60000, 2)

                            players[i]["KP"] = round((players[i]["Kill"]+players[i]["Assist"])/max(game[1]["teams"][i]["championsKills"], 1), 4)

                            players[i]["Gold"] = data_end[lane-1 + 5*i]["totalGold"]

                            players[i]["Gold_per_min"] = round(players[i]["Gold"]/total_time*60000, 1)

                            players[i]["Gold_share"] = round(players[i]["Gold"]/game[1]["teams"][i]["totalGold"], 4)

                            players[i]["Damage_to_champions"] = data_end[lane-1 + 5*i]["stats"][38]["value"]

                            players[i]["Damage_to_champions_per_gold"] = round(players[i]["Damage_to_champions"]/players[i]["Gold"], 1)

                            players[i]["Damage_to_champions_per_min"] = round(players[i]["Damage_to_champions"]/total_time*60000, 1)

                            players[i]["Damage_to_buildings"] = data_end[lane-1 + 5*i]["stats"][48]["value"]

                            players[i]["Damage_to_buildings_per_gold"] = round(players[i]["Damage_to_buildings"]/players[i]["Gold"], 1)

                            players[i]["Damage_to_buildings_per_min"] = round(players[i]["Damage_to_buildings"]/total_time*60000, 1)

                            players[i]["Damage_to_objectives"] = data_end[lane-1 + 5*i]["stats"][50]["value"]

                            players[i]["Damage_to_objectives_per_gold"] = round(players[i]["Damage_to_objectives"]/players[i]["Gold"], 1)

                            players[i]["Damage_to_objectives_per_min"] = round(players[i]["Damage_to_objectives"]/total_time*60000, 1)

                            players[i]["Damage_shielded"] = data_end[lane-1 + 5*i]["stats"][47]["value"]

                            players[i]["Damage_shielded_per_gold"] = round(players[i]["Damage_shielded"]/players[i]["Gold"], 2)

                            players[i]["Damage_shielded_per_min"] = round(players[i]["Damage_shielded"]/total_time*60000, 1)

                            players[i]["Healing"] = data_end[lane-1 + 5*i]["stats"][52]["value"]

                            players[i]["Healing_per_gold"] = round(players[i]["Healing"]/players[i]["Gold"], 2)

                            players[i]["Healing_per_min"] = round(players[i]["Healing"]/total_time*60000, 1)

                            players[i]["Crowd_control"] = data_end[lane-1 + 5*i]["stats"][51]["value"]

                            players[i]["Crowd_control_gold"] = round(players[i]["Crowd_control"]/players[i]["Gold"], 1)

                            players[i]["Crowd_control_per_min"] = round(players[i]["Crowd_control"]/total_time*60000, 1)

                            players[i]["Damage_taken"] = data_end[lane-1 + 5*i]["stats"][42]["value"]

                            players[i]["Damage_taken_gold"] = round(players[i]["Damage_taken"]/players[i]["Gold"], 1)

                            players[i]["Damage_taken_per_min"] = round(players[i]["Damage_taken"]/total_time*60000, 1)

                            players[i]["Vision"] = data_end[lane-1 + 5*i]["stats"][33]["value"]

                            players[i]["Vision_per_min"] = round(players[i]["Vision"]/total_time*60000, 1)
                            
                            players[i]["Game_length"] = total_time/60000
                            
                            for team in match["teams"]:
                                
                                if team["id"] == players[i]["Player_team"]:
                                    
                                    players[i]["Result"] = 1 if team["result"]["outcome"] == 'win' else 0
                            if 'Result' not in players[i].keys():
                                players[i]['Result'] = None
                                print(match["teams"], players[i]["Player_team"], mapping["esportsGameId"])
                            
                        for i in range(2):

                            players[i]["GD_15"] = round(players[i]["Gold_15"]/(players[i]["Gold_15"] + players[(i+1)%2]["Gold_15"]), 4)

                            players[i]["XPD_15"] = round(players[i]["XP_15"]/(players[i]["XP_15"] + players[(i+1)%2]["XP_15"]), 4)

                            players[i]["CSD_15"] = round(players[i]["CS_15"]/max((players[(i+1)%2]["CS_15"] + players[i]["CS_15"]), 1), 4)

                            players[i]["GD"] = round(players[i]["Gold"]/(players[i]["Gold"] + players[(i+1)%2]["Gold"]), 4)

                        for i in range(2):
                            for player in players_obj:
                                if players[i]["Player_id"] == player["player_id"] and players[i]['Result'] != None:
                                    player["match_history"].append(players[i])

with open('players_record.json', 'w') as f:
    json.dump(players_obj, f, indent = 4)
            
```

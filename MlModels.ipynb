```python
import pandas as pd
import json, os
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
def get_events_from_file(file_name, path):
    # Load the JSON file
    file = os.path.join(path, file_name)
    with open(file, 'r') as file:
        data = json.load(file)

    # Initialize an empty list to store the data you want
    selected_data = []

    # Iterate over the JSON data and extract the specific columns
    for i in range(len(data['intervaled_events'])):
        for item in data['intervaled_events'][i]:

            if item['eventType'] == 'game_end':
                y = item['teamID']
            
            else:
                selected_data.append({
                    'eventType': item['eventType'],
                    'teamID': item['teamID'] if 'teamId' in item else None  # Handle cases where 'teamId' might be missing
                })

    # Create a DataFrame from the selected data
    x = pd.DataFrame(selected_data)

    # Now df contains the specific columns from the JSON data
    return x, y
```

```python
cwd = os.path.abspath(os.getcwd())
path = os.path.join(cwd, 'intervals/')
X = []
Y = []
for file in os.listdir(path):
    x, y = get_events_from_file(file, path)
    X.append(x)
    Y.append(y)

# Concatenate all the DataFrames in the list
print(X)
print(Y)
```

```python
# Split the data into a training set and a testing set
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
# Step 3: Create and train the Random Forest model
# You can adjust the hyperparameters as needed
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)

# Step 4: Evaluate the model's performance on the testing data
y_pred = rf_model.predict(X_test)

# Calculate the accuracy of the model
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.2f}")
```

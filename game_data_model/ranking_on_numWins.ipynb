```python
import pandas as pd
import numpy as np
```

```python
# read in the data
df = pd.read_csv('all_data_1.csv')
print(df.head())
```

```python
# Initialize team ratings (Elo ratings)
ratings = {team: 1500 for team in df['teamId']}

# Elo rating parameters
K = 32  # Elo K-factor, which controls the magnitude of rating updates
initial_rating = 1500  # Initial rating for new teams

# Define a function to update Elo ratings
def update_ratings(winner, loser):
    p_winner = 1 / (1 + 10 ** ((ratings[loser] - ratings[winner]) / 400))
    p_loser = 1 - p_winner

    ratings[winner] = ratings[winner] + K * (1 - p_winner)
    ratings[loser] = ratings[loser] - K * p_loser

# Update ratings based on game outcomes
for i, row in df.iterrows():
    winner = df['teamId'][i] if row['outcome'] == 1 else df['teamId'][(i + 1) % 2]
    loser = df['teamId'][(i + 1) % 2]

    update_ratings(winner, loser)

# Sort teams by their ratings
sorted_teams = [team for team, rating in sorted(ratings.items(), key=lambda x: x[1], reverse=True)]

print("Team Rankings:")
for rank, team in enumerate(sorted_teams, start=1):
    print(f"Rank {rank}: {team}")

# Create a dataframe to store the rankings
rankings_df = pd.DataFrame(columns=['teamId', 'rank'])
rankings_df['teamId'] = sorted_teams
rankings_df['staisticalRank'] = rankings_df.index + 1

# Save the rankings to a CSV file
rankings_df.to_csv('statistical_rankings.csv', index=False)
```
